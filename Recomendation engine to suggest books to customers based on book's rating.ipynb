```python
#Importing all basic libraries
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import scipy
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading data
Data = pd.read_csv("Book.csv", encoding='latin1')
Data=Data.drop(['Unnamed: 0'],axis=1)
Data
```

```python
#renaming columns
Data= Data.rename(columns = {"User.ID" :"ID", "Book.Title" :"Title", "Book.Rating": "Rating"})
Data
```

```python
#Sorting dataset by USERID
Data.sort_values(by=['ID'])
```

```python
#no. of unique users in the dataset
len(Data.ID.unique())
```

```python
#No of Books in dataset
len(Data.Title.unique())
```

```python
#Creating pivot table b/w unique users and unique books
user_Data = Data.pivot_table(index='ID',columns='Title',values='Rating').reset_index(drop=True)
user_Data
```

```python
#apply imputation
user_Data.fillna(0,inplace = True)
user_Data
```

```python
#importing functions for cosine distance
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
#Calculating cosin distance
User_sim = 1 - pairwise_distances(user_Data.values,metric='cosine')
User_sim
```

```python
#store results
User_sim_df = pd.DataFrame(User_sim)
User_sim_df
```

```python
User_sim_df.iloc[0:10,0:10]
```

```python
#most similar users
User_sim_df.idxmax(axis=1)[0:]
```

```python
#IBCF
#Fetching ratings for Clara Callan
Clara_Callan_rating = user_Data['Classical Mythology']
```

```python
#finding the correlation with different books
similar_to_Clara= user_Data.corrwith(Clara_Callan_rating)
```

```python
corr_Clara = pd.DataFrame(similar_to_Clara,columns=['Correlation'])
corr_Clara.dropna(inplace=True)
corr_Clara.head()
                  
```
